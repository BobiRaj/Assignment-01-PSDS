### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans. Matplotlib is a Python library used for data visualization and creating various types of charts and plots. It is widely used in scientific computing, data analysis, and machine learning. Matplotlib provides a wide range of options to customize plots, including line styles, markers, colors, labels, and annotations. 

Some of the key features of Matplotlib are:

- It is easy to get started with and offers a lot of flexibility in terms of customization.
- It supports a wide range of data visualization techniques, including line charts, scatter plots, bar plots, histograms, and many others.
- It is open-source and has a large community of developers, which means there is a lot of documentation and resources available for users.

Five types of plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line chart: A line chart is a type of chart that displays data as a series of points connected by straight lines. It is often used to show trends or changes over time.

2. Scatter plot: A scatter plot is a type of chart that displays data as a collection of points. It is often used to show the relationship between two variables.

3. Bar chart: A bar chart is a type of chart that displays data as a series of bars. It is often used to compare data across categories.

4. Histogram: A histogram is a type of chart that displays the distribution of a dataset. It is often used to show the frequency distribution of continuous data.

5. Pie chart: A pie chart is a type of chart that displays data as a circle divided into segments. It is often used to show the proportion of data in each category.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Ans. A scatter plot is a type of plot or chart that is used to display the relationship between two variables. It is a two-dimensional graph where each point represents a value for two different variables. The variables are plotted along the x and y axes, and a point is placed at the intersection of the two values. Scatter plots are often used to identify patterns and relationships between the variables, and to detect outliers or unusual values in the data. They are commonly used in fields such as statistics, data science, and machine learning.

```python
x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.xlabel("x lable")
plt.ylabel('y label')
plt.title('Scatter plot')
# plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(4, figsize=(8,6))

axs[0].plot(x,y, color='red', label='fig_1')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend()

axs[1].plot(x1,y1, color='blue', label='fig_2')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend()

axs[2].plot(x2,y2, color='green', label='fig_3')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].legend()

axs[3].plot(x3,y3, color='pink', label='fig_4')
axs[3].set_xlabel('x')
axs[3].set_ylabel('y')
axs[3].legend()

plt.subplots_adjust(hspace=0.6)

fig.suptitle('4 subplots figure')

plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans. A bar plot, also known as a bar chart, is a type of plot that represents categorical data with rectangular bars. The height or length of each bar is proportional to the value it represents. Bar plots are used to compare values across different categories or to show the frequency of different categories.

Bar plots can be used to display various types of data, such as numerical data, discrete data, or data that can be grouped into categories. They are commonly used in business, finance, economics, and other fields to represent data in a clear and concise manner. They can also be used for exploratory data analysis, data visualization, and data communication.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel('Company name')
plt.ylabel('Profit')
plt.title('Company profit comparsion')
plt.show()
```
